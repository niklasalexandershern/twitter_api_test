```python
import tweepy
import pandas as pd
import configparser
```

```python
#read configs
config = configparser.ConfigParser()
config.read('config.ini')

consumer_key = config['twitter']['api_key']
consumer_secret = config['twitter']['api_key_secret']
access_token = config['twitter']['access_token']
access_token_secret = config['twitter']['access_token_secret']
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token,access_token_secret)
api = tweepy.API(auth)
```

```python
cursor = tweepy.Cursor(api.search_tweets,q="#j√∏depas",tweet_mode="extended").items(10)
```

```python
number_of_tweets = 200
tweets = []
likes = []
time = []

for i in tweepy.Cursor(api.user_timeline,id="Erhvervs_min",tweet_mode="extended").items(number_of_tweets):
    tweets.append(i.full_text)
    likes.append(i.favorite_count)
    time.append(i.created_at)
```

```python
df = pd.DataFrame({'tweets':tweets,'likes':likes,'time':time})
```

```python
df = df.applymap(str)
```

```python
df.to_excel('output.xlsx')
```
